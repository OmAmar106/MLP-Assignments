From the MNIST dataset, create a new dataset by taking first 10000 images for training and next 2000 images for testing.

    Collect all digit-5 (Positive class) and digit-3 (Negative class) images and stack them properly as a single datamatrix.
    By convention, keep all digit-5 images from index 0 to i followed by digit-3 images from index i+1 to n (i denotes the end index of digit-5 images)
    Similarly, collect the respective labels and store it in a variable (Do sanity check).
    Set the label values to 1 for positive class and -1 for negative class.
    Instantiate a perceptron classifier with the below parameters
        random_state = 42
        learning rate = 1
        Train for 100 iterations (epoch)
        Should shuffle the dataset for each epoch (iteration)
        Include the intercept (bias) term.
        No regularization

Use the training set for fitting the perceptron model and use the test data to make the predictions. Answer the below questions.

```python
from sklearn.datasets import fetch_openml
from sklearn.linear_model import Perceptron

X = fetch_openml('mnist_784')
```

```python
X,y = X.data,X.target
```

```python
X.head()
```

```python
y.head()
```

```python
X_train = X[:10000]
X_test = X[10000:120000]

Y_train = y[:10000]
Y_test = y[10000:120000]
```

```python
X_train_final = X_train[Y_train=='5']
import pandas as pd
X_train_concat = pd.concat([X_train_final,X_train[Y_train=='3']],ignore_index=True)
Y_train_concat = pd.concat([Y_train[Y_train=='5'],Y_train[Y_train=='3']],ignore_index=True)
```

```python
X_test_concat = pd.concat([X_test[Y_test=='5'],X_test[Y_test=='3']],ignore_index=True)
Y_test_concat = pd.concat([Y_test[Y_test=='5'],Y_test[Y_test=='3']],ignore_index=True)
```

```python
Y_train_concat.replace('5',1,inplace=True)
Y_train_concat.replace('3',-1,inplace=True)
```

```python
Y_test_concat.replace('5',1,inplace=True)
Y_test_concat.replace('3',-1,inplace=True)
```

```python
print(Y_train_concat[Y_train_concat==1].shape)
print(Y_train_concat[Y_train_concat==-1].shape)
```

```python
print(X_train_concat.shape,Y_train_concat.shape)
Y_train_concat = Y_train_concat.astype(int)
```

```python
per = Perceptron(
    random_state=42,
    eta0=1,
    max_iter=100,
    shuffle=False,
    fit_intercept=True,
    penalty=None
)
per.fit(X_train_concat,Y_train_concat)
```

```python
Y_predict = per.predict(X_test_concat)
```

```python
assert(Y_predict.shape==Y_test_concat.shape)
from collections import Counter
import numpy as np
Y_predict = pd.Series(map(int,Y_predict))
print(Counter(Y_predict))
```

```python
Y_test_concat = np.array(Y_test_concat).astype(int).flatten()
Y_predict = np.array(Y_predict).astype(int).flatten()
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test_concat,Y_predict))
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score

def func(a,b):
    print("Accuracy Score : ",accuracy_score(a,b))
    print("Precision Score : ",precision_score(a,b))
    print("Recall Score : ",recall_score(a,b))

print(func(Y_test_concat,Y_predict))
```
