```python
import seaborn as sns
from copy import deepcopy
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df.head(5)
```

```python
set1 = deepcopy(df[['ID','RecipeNumber','RecipeCode','UserReputation','ThumbsUpCount','ThumbsDownCount','BestScore','Rating']])
hm = set1.corr(method='pearson')
sns.heatmap(hm,annot=True) # with annot , only the color is there , no numbers
```

```python
sns.scatterplot(x=df['RecipeNumber'],y=df['RecipeCode'],hue=df['Rating']) #hue means the darkness
```

```python
sns.scatterplot(x=df['ThumbsUpCount'],y=df['BestScore'],hue=df['Rating'])
```
