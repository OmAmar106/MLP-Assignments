```python
x = [2,5,10]
y = [1,2,3]
plt.plot(x,y,'ks:')
plt.show()
```

1. 'k': Specifies the color (black).
2. 'o': Specifies the marker style (circle).
3. '--': Specifies the line style (dashed).

Alternatives and Their Meanings:

1. Color Codes:
   'b': Blue
   'g': Green
   'r': Red
   'c': Cyan
   'm': Magenta
   'y': Yellow
   'k': Black
   'w': White

2. Marker Styles:
   'o': Circle
   's': Square
   '^': Upward-pointing triangle
   'v': Downward-pointing triangle
   '<': Leftward-pointing triangle
   '>': Rightward-pointing triangle
   '*': Star
   '+': Plus
   'x': Cross
   'd': Diamond
   'h': Hexagon

3. Line Styles:
   '-': Solid line
   '--': Dashed line
   '-.': Dash-dot line
   ':': Dotted line

Combining Examples:
'ro-': Red solid line with circular markers.
'g^--': Green dashed line with upward-pointing triangle markers.
'b*:': Blue dotted line with star markers.
'ms-.': Magenta dash-dot line with square markers.

```python
y2 = [3,4,5]

plt.plot(x,y,'ko--',label='eq1')
plt.plot(x,y2,'bs:',label='eq2')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph')
plt.legend() # this adds the label that we defined above 
plt.grid(True,c='b') # in this , k always indicates black (since b is blue)
plt.show()
```

```python
import numpy as np

x = np.random.randint(1,100,size=20)
y = np.random.randint(1,20,size=20)
```

```python
plt.scatter(x,y,c='k',marker='s',alpha=0.6,s=20) # s is the size of the points in the graph
# alpha is the transparency of the points in the graph
plt.xlabel('x')
plt.ylabel('y')
plt.title('graph')
```

```python
x = np.random.normal(20,1,100)
x
```

```python
plt.hist(x,bins=13)
plt.title('Graph',fontsize=25)
plt.show()
```

```python
names = ['A','B','C']
vals = [10,11,9]

plt.bar(names,vals,color='skyblue',width=0.7)
plt.xlabel("Names")
plt.ylabel("Values")
plt.show()
```

```python
vals = [20,30,50]
plt.pie(vals,labels=names,explode=[0,0.05,0],autopct="%.1f",startangle=90,shadow=True) # explode takes out some part of it
plt.title('Pie Chart')
plt.show()
```

```python
x = np.arange(100)
P1 = np.cos(x)
P2 = np.sin(x)
plt.subplot(221) # in the number inside it , the first number represents the number of rows, second represent the number of columns and third represents the position of subplots 
plt.plot(P1)
plt.title('sin')
plt.subplot(222)
plt.plot(P2)
plt.title('cos')
plt.show()
```

```python
fig,axis = plt.subplots(2,2)
axis[0,0].plot(np.sin(np.arange(100)))
axis[1,1].bar(['1'],[2])
plt.show()
```

Usnig subplots , we can create many sub graphs

```python
ax = plt.axes(projection='3d')
X = np.random.randint(1,100,size=100)
Y = np.random.randint(1,100,size=100)
Z = np.random.randint(1,100,size=100)
ax.scatter(X,Y,Z)
plt.show()
```
